```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import random

def ufo_data_gen(background):

  # enable generating infinite amount of batches
  # bg_name = 'background.jpg'
  # background = Image.open(bg_name)
  # ufo_pil = np.asarray(ufo_pil) / 255.
  bg_x, bg_y, _ = np.asarray(background).shape
  print(bg_x, bg_y)
  for item in range(10):
    img_name = 'train_image_{}'.format(item)
    print('Generating {}'.format(img_name))
    ufo_count = 0
    fresh_bg = background.copy()
    fresh_bg.show()

    while ufo_count < 4:
      ufo_num = np.random.randint(0,9)
      ufo_path = 'classifier/numbers/{}/{}_{}.jpg'.format(ufo_num, ufo_num, np.random.randint(1,200))
      ufo_pil = Image.open(ufo_path)
      ufo_pil = random_scale(ufo_pil)
      # ufo_pil = np.asarray(ufo_pil) / 255.
      # print(ufo_pil.shape)
      x1 = np.random.randint(100, bg_y - 100)
      y1 = np.random.randint(100, bg_x - 200)
      # print(x1,y1)

      fresh_bg.paste(ufo_pil, (x1, y1))

      ufo_count += 1
    
    fresh_bg.save('yolo_train_data/' + img_name + '.jpg')

bg_name = 'spacy/background.jpg'
background = Image.open(bg_name)
fresh_bg = background.copy()
fresh_bg.show()

ufo_data_gen(background)
```
